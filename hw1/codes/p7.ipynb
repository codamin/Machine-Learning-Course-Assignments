```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import mean_squared_error as mse

%matplotlib inline
```

```python
m = 100
x = np.linspace(0, 2, 100)
x = x.reshape(m, 1)
y = np.cos(2 * np.pi * x) + 0.6 * np.random.randn(100, 1)
y = y.reshape(m, 1)
```

### plotting the data

```python
plt.scatter(x, y);
```

### Applying Linear Regression with different degrees
#### Meanwhile we find the mean squared errors and 

```python

degrees = [1, 4, 8, 12, 15, 20]
fig, axs = plt.subplots(3,2, figsize=(15,15))
axs = axs.flatten()

for i, deg in enumerate(degrees):
    poly_features = PolynomialFeatures(degree=deg, include_bias=False)
    X_poly = poly_features.fit_transform(x)

    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)
    b, w = lin_reg.intercept_, lin_reg.coef_
    y_fitted = np.dot(X_poly, w.T) + b
    
    MSE = mse(y, y_fitted)
    
    print('degree = ', deg, ':')
    print('    mse = ', MSE)
    print('    bias = ', np.mean(y_fitted - y))
    print('    variance = ', np.var(y_fitted - y))
    
    axs[i].plot(x, y_fitted, 'r-', label='prediction')
    axs[i].plot(x, y, 'b.', label='observation')
    axs[i].set_ylim(-2, 2)
    axs[i].set_xlabel('x')
    axs[i].set_ylabel('cos(x)')
    axs[i].set_title('polynomial degree = ' + str(deg))
    axs[i].legend()
```

As it can be seen, corresponding model to degrees 1 and 4 are underfitted.
Degree 8 is fitted.
Degrees 15 and 20 are overfitted because its plot is easily change by a few points of data and hence it has a very fluctuating shape.

Also as you can see, in higher degrees, since the shape is highly fitted to the training data, variance is smaller.
